```python
# Importing necessary libraries
from transformers import MarianMTModel, MarianTokenizer
import gradio as gr

# Defining the model name for English -> French
model_name = 'Helsinki-NLP/opus-mt-en-fr'

# Loading the tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate(text):
    # Tokenizing the input text
    tokenized_text = tokenizer([text], return_tensors='pt')
    
    # Performing the translation
    translation = model.generate(**tokenized_text)
    
    # Decoding the translated text
    translated_text = tokenizer.decode(translation[0], skip_special_tokens=True)
    
    return translated_text

# Creating a Gradio interface
iface = gr.Interface(
    fn=translate, 
    inputs=gr.Textbox(label="Enter text to translate"), 
    outputs=gr.Textbox(label="Translated text"),
    title="English to French Simple Translator",
    description="Enter English text and get the French translation."
)

# Launching the Gradio interface
iface.launch()
```
